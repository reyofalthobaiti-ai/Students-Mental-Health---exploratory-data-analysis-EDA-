```python
students_df.Gender.value_counts()
```

```python
#it appears that females have a higher rate of depression!
students_df.groupby('Gender')['Depression'].value_counts()
```

```python
#number of courses taken by females
females_courses=students_df.groupby('Gender')['Course'].value_counts().loc['Female']
females_courses
```

```python
#analyzing if course relates to depression
students_df.groupby(students_df['Course']=='Engineering')['Depression'].value_counts()
```

```python
#1- filtering only females students 2- groupby course+depression 
students_df[students_df.Gender=='Female'].groupby(students_df['Course'])['Depression'].value_counts()
```

```python
#its clear that the females who study engineering are the most depressed.
result=students_df[students_df.Gender=='Female'].groupby(students_df['Course'])['Depression'].value_counts()
result.idxmax()
```

```python
result[result.index.get_level_values('Depression') == 1].plot(kind='bar',title='Female Depression By Course',figsize=(13,8))
```

```python
students_df['PanicAttack'].value_counts()
```

```python
#it appears that anxiety is much higher than panic attack
students_df['Anxiety'].value_counts()
```

```python
#as shown females also have higher anxiety
students_df.groupby('Gender')['Anxiety'].value_counts()
```

```python
#lets visualize it
students_df.groupby('Gender')['Anxiety'].value_counts().unstack().plot(kind='bar',title='Anxiety by Gender',figsize=(12,6),color=('steelblue','salmon'))
```

```python
students_df.groupby('CGPA')['Anxiety'].value_counts()
```

```python
# As visualized, students with higher GPAs tend to report higher anxiety levels,
# suggesting high achievers may experience more academic pressure
students_df.groupby('Anxiety')['CGPA'].plot(kind='hist',title='GPA by Anxiety',figsize=(12,8),legend=True,stacked=True)
plt.xlabel('gpa')
plt.ylabel('anxiety count')
```
