```python
#The distribution shows a higher number of male students compared to female students in the dataset
sns.displot(students_df,x='gender',binwidth=1,hue='gender')
plt.title('Gender Distribution of Students')
plt.show()
```

```python
students_df['gender'].value_counts()
```

```python
# The plot shows the distribution of students across degree majors, 
# with Data Science appearing as the most common major in the dataset.
sns.displot(students_df,x='degree_major',hue='degree_major',height=4,aspect=2)
plt.title("Distribution of Students by Degree Major")
plt.show()
```

```python
# The plot shows that the CGPA ranges 3.00–3.49 and 3.50–4.00 have similar counts of students
sns.displot(students_df,x='cgpa',hue='cgpa',height=3,aspect=2)
plt.title('CGPA distribution')
```

```python
# The visualization indicates that many students in the dataset do not exercise regularly
# which could be explored further in relation to mental health variables
sns.displot(students_df,x='sports_engagement',hue='sports_engagement',height=3,aspect=2.5)
plt.title('Sports Engagement')
```

```python
students_df
```

```python
students_df['cgpa'] = students_df['cgpa'].str.split('-').str[0].astype(float)
```

```python
# Software Engineering students have the highest average CGPA (~3.5)
# compared to all other degree majors, suggesting stronger academic
# performance or potentially less rigorous grading in that field
sns.displot(data=students_df,x='degree_major',y='cgpa',height=3,aspect=2.8,hue='degree_major')
plt.title('Academic Performance Across Degree Majors')
```

```python
# Students with moderate sports engagement (~medium level) tend to have 
# the highest average CGPA (~3.4), suggesting a balance between physical 
# activity and academic focus. Excessive or no sports engagement is 
# associated with slightly lower academic performance.
students_df.groupby('sports_engagement')['cgpa'].mean().plot(kind='bar',figsize=(12,4),color=['steelblue', 'salmon', 'mediumpurple', 'gold'])
```

```python
# Students with higher study satisfaction tend to have higher CGPAs,
# suggesting that engagement and enjoyment in academics may positively 
# influence academic performance — though causality cannot be confirmed.
sns.barplot(data=students_df, x='study_satisfaction', y='cgpa')
```

```python
sns.boxplot(data=students_df,x='anxiety',y='cgpa')
```

```python
plt.figure(figsize=(12,6))
sns.kdeplot(data=students_df, x='academic_pressure', hue='anxiety', fill=True, alpha=0.5)
plt.title('Academic Pressure Distribution by Anxiety')
```

```python
# Religious activities appear to be the most common stress relief method
# among students, suggesting that spirituality plays a significant role
# in how students cope with academic pressure
top10 = students_df['stress_relief_activities'].value_counts().head(10)
top10.plot(kind='barh', figsize=(12,6))
plt.title('Top 10 Stress Relief Activities')
```

```python
# Multivariate analysis reveals interconnected mental health patterns:
# students with higher financial concerns and future insecurity tend to 
# report higher depression and anxiety levels, while CGPA shows a modest
# positive correlation with academic workload — suggesting high achievers
# take on more pressure despite the mental health cost
sns.pairplot(students_df[['cgpa', 'financial_concerns' ,'academic_workload ','anxiety', 'future_insecurity','depression']], 
             hue='depression')
```
